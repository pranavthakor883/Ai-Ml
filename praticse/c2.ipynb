```python
df = pd.get_dummies(df,columns=["city"],drop_first=True,dtype=int)
```

```python
df["loyalty_member"]=df["loyalty_member"].replace({"Yes":0,"No":1})
```

```python
df["gender"]=df["gender"].replace({"Male":0,"Female":1})
```
